## üß© Population and Sample

**Concept:**
- **Population** ‚Üí Whole group you want to study.  
- **Sample** ‚Üí Small part taken from that group to study easily.

**Example:**
Your college has 2,000 students (population).  
You survey 100 students about screen time (sample).

**Scenario:**
Instead of asking everyone, you ask a few and use their answers to guess for all.

**Key Point:**
A good sample should represent the whole population fairly.

---

## üìä Descriptive vs Inferential Statistics

**Concept:**
- **Descriptive Statistics** ‚Üí Used to **describe** data you already have (average, percentage, chart).  
- **Inferential Statistics** ‚Üí Used to **predict or conclude** about a larger group based on a sample.

**Example:**
If 100 students‚Äô average screen time = 4 hrs ‚Äî  
- Saying ‚ÄúAverage screen time is 4 hrs‚Äù ‚Üí **Descriptive**  
- Saying ‚ÄúAll college students likely spend around 4 hrs‚Äù ‚Üí **Inferential**

**Scenario:**
Descriptive = Summary of known data.  
Inferential = Guess for the whole group.

**Key Point:**
Descriptive = ‚ÄúWhat is happening?‚Äù  
Inferential = ‚ÄúWhat might happen?‚Äù

# üìä Measures of Central Tendency

These are used to find the **center or average** of a dataset.  
The three main measures are **Mean**, **Median**, and **Mode**.

---

## üí° Concept
- **Mean** ‚Üí Arithmetic average  
- **Median** ‚Üí Middle value when data is sorted  
- **Mode** ‚Üí Most frequent value  

These help us understand where most data values lie.

---

## üìò Example
Marks of 5 students: 50, 60, 70, 80, 90

- **Mean** = (50 + 60 + 70 + 80 + 90) / 5 = 70  
- **Median** = 70 (middle value)  
- **Mode** = None (no repeated value)

---

## üåç Scenario
A teacher wants to know the **average marks** of the class: uses **Mean**.  
Wants to find the **middle performer**: uses **Median**.  
Wants to know the **most common mark**: uses **Mode**.

---

## üß† Key Points
- **Mean** is affected by extreme values (outliers).  
- **Median** is better when data has outliers.  
- **Mode** is useful for categorical dat brand).

# üìä Measures of Variability

**Concept:**  
These show how much data values differ from each other or from the average.  
In simple terms ‚Äî they measure **how spread out** the data is.

---

### **1Ô∏è‚É£ Range**
Shows how spread out the data is.  
**Formula:** `Range = Max ‚àí Min`  
**Example:** Ages 20, 25, 40 ‚Üí Range = 40 ‚àí 20 = **20**

---

### **2Ô∏è‚É£ Mean Absolute Deviation (MAD)**
Shows the **average distance** of each data point from the mean.  
**Formula:**  
\[
MAD = \frac{\sum |x - \bar{x}|}{n}
\]  
**Example:**  
If all ages are close to the mean, MAD is small ‚Üí data is consistent.

---

### **3Ô∏è‚É£ Variance**
Shows how far data points are from the mean **on average (squared)**.  
**Formula:**  
\[
Variance = \frac{\sum (x - \bar{x})^2}{n}
\]  
**Example:**  
Larger variance = data is more spread out.

---

### **4Ô∏è‚É£ Standard Deviation (SD)**
It‚Äôs the **square root of variance**, showing spread in **original units**.  
**Formula:**  
\[
SD = \sqrt{Variance}
\]  
**Example:**  
If average age = 40 and SD = 5 ‚Üí most people are between **35 and 45**.

---

### **üí° Key Points**
- **Range** ‚Üí simple spread  
- **MAD** ‚Üí average difference  
- **Variance** ‚Üí squared spread  
- **SD** ‚Üí real-world spread (most used)


````python
# Minimum and Maximum Credit Limit
min_r = dataset['Credit_Limit'].min()
max_r = dataset['Credit_Limit'].max()

print("Minimum Credit Limit:", min_r)
print("Maximum Credit Limit:", max_r)

# Range
range_r = max_r - min_r
print("Range of Credit Limit:", range_r)

````

````python
# MAD
A = np.array([75, 65, 70, 80, 60])
B = np.array([90, 85, 100, 95, 110])

mad_A = np.sum(np.abs(A - np.mean(A))) / len(A)
mad_B = np.sum(np.abs(B - np.mean(B))) / len(B)

print("MAD of A:", mad_A)
print("MAD of B:", mad_B)
````

### üìä Percentiles and Quartiles

**Percentiles** divide a dataset into **100 equal parts**.  
Each percentile indicates the value **below which a given percentage of observations** fall.  
For example, the **90th percentile** is the value below which **90% of the data** lies.

**Quartiles** are a special case of percentiles that **divide data into 4 equal parts**:

- **Q1 (First Quartile / 25th Percentile)** ‚Üí 25% of data lies below this value.  
- **Q2 (Second Quartile / 50th Percentile / Median)** ‚Üí 50% of data lies below this value.  
- **Q3 (Third Quartile / 75th Percentile)** ‚Üí 75% of data lies below this value.  

Together, quartiles help identify the **spread and distribution** of data.  
The difference between Q3 and Q1 is called the **Interquartile Range (IQR)**, which measures the **middle 50% spread** of the dataset.


````python
# üéØ Calculating percentiles for Customer_Age

min_age = np.percentile(dataset["Customer_Age"], 0)   # 0th Percentile (Minimum Age)
q3_age = np.percentile(dataset["Customer_Age"], 75)   # 75th Percentile (Q3)

print("Minimum Age (0th Percentile):", min_age)
print("Third Quartile (75th Percentile):", q3_age)

````

### üì¶ Detecting and Removing Outliers using Boxplot

Using a **boxplot** is the **best visual way** to detect **outliers** in your dataset.


````python
sns.boxplot(x="Customer_Age", data=dataset)
plt.title("Boxplot of Customer Age")
plt.show()
````

````python
# Calculate Q1 and Q3
Q1 = dataset["Customer_Age"].quantile(0.25)
Q3 = dataset["Customer_Age"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filter the dataset
dataset_no_outliers = dataset[(dataset["Customer_Age"] >= lower_limit) & (dataset["Customer_Age"] <= upper_limit)]

print("Original dataset shape:", dataset.shape)
print("After removing outliers:", dataset_no_outliers.shape)
````

````python
plt.figure(figsize=(10, 5))

# Before removal
plt.subplot(1, 2, 1)
sns.boxplot(x="Customer_Age", data=dataset)
plt.title("Before Outlier Removal")

# After removal
plt.subplot(1, 2, 2)
sns.boxplot(x="Customer_Age", data=dataset_no_outliers)
plt.title("After Outlier Removal")

plt.tight_layout()
plt.show()
````

### üí• What is an Outlier?

An **outlier** is a data point that is **significantly different** from other observations in a dataset.  
It either lies **much higher or much lower** than most of the other values.

For example:  
If most customers are aged between **25 and 50**, but one customer is **95**, that value (95) is an **outlier**.

---

### üéØ Why Do We Detect and Handle Outliers?

Outliers can:
- **Distort averages (mean):** A few extreme values can make the average misleading.  
- **Affect machine learning models:** Algorithms like regression or clustering are sensitive to extreme points.  
- **Create wrong insights:** Outliers may make data appear more spread out or skewed than it actually is.

So, we detect and sometimes remove outliers to make the analysis **more accurate and reliable**.

---

### üß† Real-Life Data Science Scenario

**Example:**  
A bank is analyzing the **average monthly spending** of its customers to find who might need a premium credit card.  

- Most customers spend between **‚Çπ10,000 and ‚Çπ50,000**.  
- But one corporate client spends **‚Çπ8,00,000** every month.  

That ‚Çπ8,00,000 value is an **outlier** ‚Äî it‚Äôs real, but it doesn‚Äôt represent typical customer behavior.  
If not handled properly:
- The **average spending** will rise artificially.  
- The bank might **incorrectly classify** normal customers as high-spending ones.

üëâ That‚Äôs why data scientists often **detect, analyze, and sometimes remove** outliers ‚Äî to keep results meaningful.

---

**‚úÖ In short:**  
> Outliers are extreme values that can distort your analysis.  
> Handling them correctly ensures your data-driven decisions are accurate and trustworthy.

## üé≤ Probability and Random Variables

---

### üéØ What is Probability?

**Probability** is the measure of how likely an event is to occur.  
It ranges from **0 to 1**, where:
- **0** ‚Üí The event is **impossible**  
- **1** ‚Üí The event is **certain**

#### üßÆ Formula:
\[
P(E) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}}
\]

#### üé≤ Example:
If you toss a fair coin,
- Probability of getting **Heads** = 1/2  
- Probability of getting **Tails** = 1/2

---

### üé≤ What is a Random Variable?

A **Random Variable (RV)** is a variable that represents the **numerical outcome of a random event**.  
It connects **probability** with **numbers**.

For example:
- Tossing a coin ‚Üí Let `X = 1` if Head, `X = 0` if Tail  
  ‚Üí Here, `X` is a **random variable**.

---

### üß© Types of Random Variables

#### 1Ô∏è‚É£ Discrete Random Variable
Takes **countable** values (like 0, 1, 2, 3...).  
Examples:
- Number of heads in 3 coin tosses  
- Number of customers visiting a shop per hour  
- Number of students passing an exam  

These are often modeled using **Binomial**, **Poisson**, or **Geometric** distributions.

---

#### 2Ô∏è‚É£ Continuous Random Variable
Takes **uncountable, infinite** values within a range.  
Examples:
- Height of students  
- Temperature in a city  
- Time taken to complete a task  

These are modeled using **Normal**, **Exponential**, or **Uniform** distributions.

---

### üß† Real-Life Data Science Example

Suppose you work for a **bank**:
- You can model the **number of credit card applications per day** as a **Discrete Random Variable** (count data).  
- You can model the **income of customers** as a **Continuous Random Variable** (ranges over real numbers).  

Understanding random variables helps data scientists in:
- **Predictive modeling**  
- **Simulation and sampling**  
- **Statistical inference and machine learning**

---

**‚úÖ Summary:**
| Concept | Meaning | Example |
|----------|----------|----------|
| **Probability** | Likelihood of an event | P(Heads) = 0.5 |
| **Random Variable** | Numeric outcome of a random process | X = number of Heads |
| **Discrete RV** | Countable values | No. of customers/day |
| **Continuous RV** | Infinite possible values | Temperature, height |

---

**üí° Tip:**  
Random variables bridge the gap between **theory (probability)** and **data (statistics)** ‚Äî forming the base of data science models.

# üìä Covariance & Correlation ‚Äî For Data Science

---

## üåü 1Ô∏è‚É£ What is Covariance?

**Definition:**  
Covariance tells us **how two variables move together** ‚Äî  
whether they increase or decrease **at the same time**.

\[
\text{Cov}(X,Y) = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{n-1}
\]

---

### üîπ **Interpretation:**

| Covariance Value | Meaning |
|-------------------|----------|
| Positive | When X increases, Y also increases |
| Negative | When X increases, Y decreases |
| Zero | No linear relationship |

---

### üíª **Example (Data Science Scenario):**

In a food delivery app (like Swiggy):

| Day | Orders (X) | Delivery Time (Y) |
|-----|-------------|------------------|
| 1 | 100 | 30 |
| 2 | 150 | 40 |
| 3 | 200 | 60 |

As the number of orders increases, delivery time also increases ‚Üí  
**Covariance = Positive**

---

## üåü 2Ô∏è‚É£ What is Correlation?

**Definition:**  
Correlation measures **how strong and in what direction** two variables are related.  
It‚Äôs a **standardized version of covariance**.

\[
r = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}
\]

---

### üîπ **Range:**

-1 ‚â§ **r** ‚â§ +1  

| Correlation Value | Meaning |
|--------------------|----------|
| +1 | Perfect positive correlation |
| 0 | No linear relationship |
| -1 | Perfect negative correlation |

---

## üåü 3Ô∏è‚É£ Covariance vs Correlation

| Basis | Covariance | Correlation |
|--------|-------------|-------------|
| **Meaning** | Shows how two variables move together | Shows how strongly & in what direction variables are related |
| **Scale** | Not standardized | Standardized (-1 to +1) |
| **Units** | Has units (depends on data) | Unitless |
| **Range** | -‚àû to +‚àû | -1 to +1 |
| **Interpretation** | Harder | Easier |

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------
# üßÆ Step 1: Create simple data
# -----------------------------
data = pd.DataFrame({
    'Study_Hours': [1, 2, 3, 4, 5, 6, 7],
    'Marks': [35, 40, 50, 60, 70, 75, 85]
})

print("üìä Simple Data:\n", data)

# -----------------------------
# üßÆ Step 2: Covariance
# -----------------------------
print("\nCovariance:\n", data.cov())

# -----------------------------
# üîó Step 3: Correlation
# -----------------------------
print("\nCorrelation:\n", data.corr())

# -----------------------------
# üìà Step 4: Graph (Scatter Plot)
# -----------------------------
sns.scatterplot(x='Study_Hours', y='Marks', data=data, color='teal')
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()

# -----------------------------
# üìä Step 5: Correlation Heatmap
# -----------------------------
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
````
